## Парсинг данных 

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_excel("data/dataset.xlsx",header=None)
data.head()
```

Из данного `DataFrame` можно увидеть 5 главных атрибутов. Первая колонка содержит название компетенции. Вторая - описание компетенции. Третья название раздела в компетенции. Четвёртая - значимость раздела. Пятая - это признаки раздела. В этом `DataFrame`'е остальные столбцы имеют одинаковость шаблонов 3-его, 4-ого и 5-ого столбца  

```python
# Формирование пустого DataFrame с названиями компетенций
df=pd.DataFrame({"название_компетенции_wsss":[],"описание_компетенции_wsss":[],"название_раздела_wsss":[],
                "значимость_раздела_wsss":[],"признаки_раздела_wsss":[]})

# Добавление дополнительных столбцов
for i in range(27):
    df["column_"+str(i)]=0
    
# Передача названий от пустого DataFrame к заполненому данными 
data.columns=df.columns

# На начальном этапе лучше все столбцы привести к единому типу данных, а после необходимые столбцы преобразовать
for data_type in df.columns:
    df[data_type]=df[data_type].astype(str)
```

### Алгоритм формирования структуры данных

```python
# Прохождение ипо каждой строке
for row in range(len(data)):
    # Прохождение по повторяющимся столбцам как было заявлено ранее
    for z,t,r in zip(range(2,28,3),range(3,28,3),range(4,28,3)):
        if pd.isnull(data.iloc[row][z])==True:
            break
        else:
            # Составление DataFrame с формированной структурой
            df.loc[len(df)]=np.nan
            df=df.shift()
            df.at[0,"название_компетенции_wsss"]=data.iloc[row][0]
            df.at[0,"описание_компетенции_wsss"]=data.iloc[row][1]
            df.at[0,"название_раздела_wsss"]=data.iloc[row][z]
            df.at[0,"значимость_раздела_wsss"]=data.iloc[row][t]
            df.at[0,"признаки_раздела_wsss"]=data.iloc[row][r]
            
# Удаление лишних столбцов, а именно всех, кроме после пятого столбца
df.drop(df.iloc[:,5:],inplace=True,axis=1)
df.head()
```

А именно подсчёт слов каждого поля текстовых столбцов. Но перед этим необходимо провести очистку данных от лишних символов. Это могут быть точки, запятые и прочие символы, влияющие как на крастеризацию, так и на подсчёт слов

### Очистка данных
В данном подразделе будет создана функция, которая переводит данные столбца к нижнему регистру, убирает все стопслова и лишние символы.

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')

#Function to clean a column
def clean_data(df):
    df = df.str.lower()
    df.replace(r'[^а-яА-Яa-zA-Z0-9]', ' ', regex=True, inplace=True)
    df = df.str.strip()
    stop_words = stopwords.words("russian")
    df = df.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
    return df
```

```python
df["описание_компетенции_wsss"]=clean_data(df["описание_компетенции_wsss"])
df["название_раздела_wsss"]=clean_data(df["название_раздела_wsss"])
df["признаки_раздела_wsss"]=clean_data(df["признаки_раздела_wsss"])
df.head()
```

### Новые атрибуты
Далее образуем новые атрибут, которые могут быть полезны для дальнейшей работы с датасетом. Они будут содержать в себе количество слов каждого поля текстового атрибута.

* `count_description` - содержит кол-во слов `описание_компетенции_wsss`
* `count_title` - содержит кол-во слов `название_раздела_wsss`
* `count_titlefeat` - содержит кол-во слов `признаки_раздела_wsss`

```python
df["count_description"]=np.nan
df["count_title"]=np.nan
df["count_titlefeat"]=np.nan
for row in range(len(df)):
    df["count_description"][row]=len(df["описание_компетенции_wsss"][row].split()) 
    
for row in range(len(df)):
    df["count_title"][row]=len(df["название_раздела_wsss"][row].split()) 
    
for row in range(len(df)):
    df["count_titlefeat"][row]=len(df["признаки_раздела_wsss"][row].split()) 

df["count_description"]=df["count_description"].astype(int)
df["count_title"]=df["count_title"].astype(int)
df["count_titlefeat"]=df["count_titlefeat"].astype(int)
```

В данном датасете есть атрибут, который должен быть типом `float`, а именно `значимость_раздела_wsss`. Приведём этот атрибут к необходимому типу данных

```python
df['значимость_раздела_wsss'] = pd.to_numeric(df['значимость_раздела_wsss'],errors='coerce')
df.info()
```

Поскольку в атрибуте `значимость_раздела_wsss` некоторые ячейки были формата string, то они удалились и на их место стоит поставить среднее значение этого столбца

```python
df["значимость_раздела_wsss"]=df["значимость_раздела_wsss"].fillna(float(int(df["значимость_раздела_wsss"].mean())))
df.info()
```

### Добавление функции стеминга
Стеминг необходим, чтобы минимизировать количество слов в словаре. Тем самым отсекая от слова всё, что находится после корня. Благодаря этому однокоренные слова будут лучше распознаваться алгоритмом машинного обучения 

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("russian") 

#Stemming data to improve model quality
def stemming_tokenizer(text):
    text = re.split('\W+', text)
    text = [stemmer.stem(word) for word in text]
    return text
```

## Формирование словарей данных
В данном разделе возьмём векторайзер `CountVectorizer`, который сформирует словарь, который можно выгрузить в csv файл `dictionary.csv`.

```python
import re
import csv
from sklearn.feature_extraction.text import CountVectorizer

stop_words = stopwords.words("russian")
#A vectorizer for count words in text
cvzer = CountVectorizer(stop_words = stop_words,min_df = 0.1, max_df = 0.25, tokenizer=stemming_tokenizer)
vec_dictionary = cvzer.fit_transform(df["признаки_раздела_wsss"])
#Getting words set
my_dict = cvzer.vocabulary_

#Writing vocabulary to csv
#You will need 'wb' mode in Python 2.x
with open('dictionary.csv', 'w', encoding='utf-16', newline='') as f: 
    w = csv.DictWriter(f,my_dict.keys())
    w.writeheader()
    w.writerow(my_dict)
```

### Анализ словаря
Проведём анализ словаря на основе 4 блоков компетенции. Анализ будет проводится по `вхождению ключевых слов` для каждого кластера

```python
from sklearn.cluster import KMeans
words = cvzer.get_feature_names()

#K-Means Algorithm for Determining a Cluster of Words
kmeans = KMeans(n_clusters = 4, random_state = 0)
kmeans.fit(vec_dictionary)
kmeans_fit = kmeans.fit_predict(vec_dictionary)

#Outputting grouped words
common_words = kmeans.cluster_centers_.argsort()[:,-1:-20:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
```

```python
df["cluster"]=kmeans_fit
```

### Наименование кластеров

* Кластер `1` наменование этого кластера может быть проблематично, но т.к. здесь мало уникальных слов. Но тем не менее определим его, как "Строительство и строительные технологии" из-за слов: ситуац, техник, средств, норм, применен
* Кластер `2` можно отнести к "Сфера услуг" из за таких слов, как: труд, охра, прав, защит
* Кластер `3` можно отнести к "Производство и инженерные технологии" из-за таких слов, как: процесс, технолог, инструмент, техник, чертеж, эффективн, производств

```python
#Naming clusters
d = {0:"Информационные и коммуникационные технологии", 1:"Строительство и строительные технологии",
     2:"Сфера услуг", 3:"Производство и инженерные технологии"}

df["nameed_clusters"] = df["cluster"].replace(d)
df.head()
```
